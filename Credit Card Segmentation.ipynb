    2) PURCHASE_TYPE 

```python
df.ONEOFF_PURCHASES.head()
```

```python
df.INSTALLMENTS_PURCHASES.head()
```

```python
def purchase_type(df):
    if((df.ONEOFF_PURCHASES==0)&(df.INSTALLMENTS_PURCHASES==0)):
        return 'none'
    if((df.ONEOFF_PURCHASES>0)&(df.INSTALLMENTS_PURCHASES>0)):
        return 'oneoff_installment'
    if((df.ONEOFF_PURCHASES==0)&(df.INSTALLMENTS_PURCHASES>0)):
        return 'installment'
    if((df.ONEOFF_PURCHASES>0)&(df.INSTALLMENTS_PURCHASES==0)):
        return 'oneoff'    
```

```python
df['PURCHASE_TYPE']=df.apply(purchase_type,axis=1)
```

# PCA

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
df_scaled=sc.fit_transform(df1)
```

```python
df_scaled
```

```python
from sklearn.decomposition import PCA
```

```python
a={}
N=range(3,12)
for n in N:
    pca=PCA(n_components=n)
    pca_variance=pca.fit(df_scaled)
    a[n]=sum(pca_variance.explained_variance_ratio_)
```

```python
pd.Series(a).plot()
```

```python
a
```

```python
df_pca_final=PCA(n_components=7)
df2=df_pca_final.fit_transform(df_scaled)
```

# K-Means

```python
from sklearn.cluster import KMeans
```

```python
distortions=[]
K=range(1,10)
for k in K:
    kmean_model=KMeans(n_clusters=k)
    kmean_model.fit(df2)
    distortions.append(kmean_model.inertia_)
```

```python
distortions
```

```python
plt.plot(K,distortions)
```

# 5 clusters

```python
km_5=KMeans(n_clusters=5,random_state=112)
km_5.fit(df2)
```

```python
km_5.labels_
```

```python
km_5.cluster_centers_
```

```python
color_map={0:'r',1:'b',2:'g',3:'y',4:'k'}
label_color=[color_map[l] for l in km_5.labels_]
plt.figure(figsize=(7,7))
plt.scatter(df2[:,0],df2[:,1],c=label_color)
```

```python
from sklearn.metrics import calinski_harabaz_score,silhouette_score
```

```python
calinski_harabaz_score(df2,km_5.labels_)
```

```python
silhouette_score(df2,km_5.labels_)
```
